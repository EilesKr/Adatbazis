```sql
SELECT v.USERNEV,
       COUNT(DISTINCT(szall.SZALLAS_NEV)) 'Különböző szálláshelyek száma'
FROM Vendeg v LEFT JOIN Foglalas f ON v.USERNEV = f.UGYFEL_FK
              LEFT JOIN Szoba sz ON f.SZOBA_FK = sz.SZOBA_ID
              LEFT JOIN szallashely szall ON sz.SZALLAS_FK = SZALLAS_ID
GROUP BY ROLLUP(v.USERNEV)
```

```sql
SELECT *,
       DATEDIFF(YEAR, SZUL_DAT, GETDATE()) 'Életkor',
       AVG(DATEDIFF(YEAR, SZUL_DAT, GETDATE())) OVER(ORDER BY SZUL_DAT ROWS BETWEEN 2 PRECEDING and 2 FOLLOWING)
FROM Vendeg
```

```sql
SELECT DISTINCT(v.USERNEV),
       v.NEV
FROM Vendeg v JOIN Foglalas f ON v.USERNEV = f.UGYFEL_FK
WHERE NOT EXISTS
(
    SELECT *
    FROM foglalas f2
    WHERE v.USERNEV = f2.UGYFEL_FK 
    AND f2.FELNOTT_SZAM + f2.GYERMEK_SZAM <> 2
)
```
